```python
import csv
from datetime import datetime, timedelta
import random

class Customer:
    def __init__(self, customer_id, name):
        self.customer_id = customer_id
        self.name = name

    def __str__(self):
        return f'{self.customer_id}: {self.name}'

class Order:
    def __init__(self, order_id, store_id, customer, items, prices):
        self.order_id = order_id
        self.store_id = store_id  # Store ID for the order
        self.customer = customer
        self.items = items
        self.prices = prices
        self.total_price = sum(prices)
        self.date = self.random_date()  # Randomized date and time

    def random_date(self):
        # Random date and time between 2023-01-01 and 2024-12-31
        start_date = datetime(2023, 1, 1)
        end_date = datetime(2024, 12, 31)
        delta = end_date - start_date
        random_seconds = random.randint(0, delta.total_seconds())  # Random number of seconds
        random_date = start_date + timedelta(seconds=random_seconds)
        return random_date.strftime("%Y-%m-%d %H:%M:%S")  # Format: YYYY-MM-DD HH:MM:SS

    def __str__(self):
        return f'Store ID: {self.store_id}, Order ID: {self.order_id}, Date: {self.date}, Items: {", ".join(self.items)}, Total: ${self.total_price:.2f}'

class Store:
    def __init__(self, store_id, name):
        self.store_id = store_id  # Unique store ID
        self.name = name
        self.customers = {}
        self.orders = []

    def add_customer(self, customer):
        if customer.customer_id not in self.customers:
            self.customers[customer.customer_id] = customer

    def add_order(self, order):
        self.orders.append(order)
        self.add_customer(order.customer)

    def get_customers(self):
        return list(self.customers.values())

    def get_orders(self):
        return self.orders

    def get_total_sales(self):
        return sum(order.total_price for order in self.orders)

    def __str__(self):
        return f'{self.name} (Store ID: {self.store_id}) - Customers: {len(self.customers)}, Orders: {len(self.orders)}, Total Sales: ${self.get_total_sales():.2f}'

class Corporation:
    def __init__(self, name):
        self.name = name
        self.stores = []

    def add_store(self, store):
        self.stores.append(store)

    def get_all_customers(self):
        customers = []
        for store in self.stores:
            customers.extend(store.get_customers())
        return customers

    def get_all_orders(self):
        orders = []
        for store in self.stores:
            orders.extend(store.get_orders())
        return orders

    def get_total_sales(self):
        return sum(store.get_total_sales() for store in self.stores)

    def __str__(self):
        return f'Corporation: {self.name}, Stores: {len(self.stores)}, Total Sales: ${self.get_total_sales():.2f}'

class Inventory:
    def __init__(self):
        self.items = {
            "Coke": 1.50,
            "Pepsi": 1.50,
            "Fanta": 1.20,
            "Sprite": 1.25,
            "Mountain Dew": 1.30
        }

    def get_price(self, item):
        return self.items[item]

def generate_random_orders(num_orders, store_id, customers, inventory):
    orders = []
    for i in range(num_orders):
        customer = random.choice(customers)
        items = random.sample(list(inventory.items.keys()), random.randint(1, 3))  # Random number of items per order (1-3)
        prices = [inventory.get_price(item) for item in items]
        order = Order(i + 1, store_id, customer, items, prices)
        orders.append(order)
    return orders

# Generate random names
random_names = ["John", "Emma", "Oliver", "Sophia", "Liam", "Isabella", "Noah", "Mia", "James", "Ava",
                "William", "Emily", "Benjamin", "Abigail", "Lucas", "Madison", "Henry", "Charlotte", "Alexander", "Amelia"]

# Create a corporation
corp = Corporation("Beverage Corp")

# Create 2500 stores with unique store_id (from 1 to 2500)
for i in range(2500):
    store_id = random.randint(1, 2500)  # Generate a random store_id between 1 and 2500
    store = Store(store_id, f"Store {i + 1}")
    corp.add_store(store)

# Create customers with unique IDs between 1 and 3000
customer_ids = random.sample(range(1, 3001), 100)
customers = [Customer(customer_ids[i], random.choice(random_names)) for i in range(100)]

# Create inventory
inventory = Inventory()

# Add random orders to each store (with random order count)
for store in corp.stores:
    num_orders = random.randint(5, 20)  # Random number of orders per store (between 5 and 20)
    orders = generate_random_orders(num_orders, store.store_id, customers, inventory)
    for order in orders:
        store.add_order(order)

# Display the date and time of each purchase
print(f"Report generated on: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
for store in corp.stores:
    print(store)
    for order in store.get_orders():
        print(order)
    print("\n")

# Display all customers
print("All Customers:")
for customer in corp.get_all_customers():
    print(customer)

# Display total sales for the corporation
print(f"\nTotal Sales for Corporation: ${corp.get_total_sales():.2f}")

# Create a list to save all customer transaction data
customer_transactions = []

# Collect customer transaction data
for store in corp.stores:
    for order in store.get_orders():
        customer_transactions.append([
            order.store_id,  # Store ID added
            order.order_id,  # Order ID
            order.customer.customer_id,
            order.customer.name,
            ", ".join(order.items),
            ", ".join(f"${price:.2f}" for price in order.prices),
            f"${order.total_price:.2f}",
            order.date
        ])

# Write the customer transaction data to a CSV file called "soda_customer_purchases.csv"
with open('soda_customer_purchases.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Store ID", "Order ID", "Customer ID", "Customer Name", "Items", "Prices", "Total Price", "Date"])
    for transaction in customer_transactions:
        writer.writerow(transaction)

print("\nCSV file 'soda_customer_purchases.csv' has been generated.")
```
